```python
# Import necessary libraries
import pandas as pd  # Data manipulation
import numpy as np  # Numerical operations
import matplotlib.pyplot as plt  # Plotting
import sklearn  # Machine learning tools

# Import specific functions
from pylab import rcParams  # Plot configuration
from sklearn.linear_model import LinearRegression  # Linear regression model
from sklearn.preprocessing import scale  # Data scaling
```

```python
# Enable inline plotting for Jupyter notebooks
%matplotlib inline

# Set default plot size to 5x4 inches
rcParams['figure.figsize'] = 5, 4

# Use 'ggplot' style for plots
plt.style.use('ggplot')
```

```python
# Generate 100 random values for 'rooms', scaled and shifted
# 'rooms' represents a feature, e.g., number of rooms in a dataset
rooms = 2 * np.random.rand(100, 1) + 3

# Display the first 10 entries of the 'rooms' array
rooms[0:10]
```

```python
# Calculate 'price' based on the number of rooms with added random noise
# The formula uses a base price factor (265), scales it by room count (6 * rooms), and adds random noise
price = 265 * 6 * rooms + np.abs(np.random.randn(100, 1))

# Display the first 10 entries of the 'price' array
price[0:10]
```

```python
plt.plot(rooms,price,'b.')
plt.xlabel('# of Rooms, 2024 Data')
plt.ylabel('Avarage House Price, 2024 Data')
plt.show()
```

```python
# Assign the 'rooms' data to feature variable X
X = rooms

# Assign the 'price' data to target variable Y
Y = price

# Initialize the Linear Regression model
LinReg = LinearRegression()

# Fit the Linear Regression model using the feature X and target Y
# This step finds the best-fitting line by minimizing the difference between predicted and actual values
LinReg.fit(X, Y)

# Calculate the coefficient of determination (R²)
# R² indicates how well the model explains the variability of the response data
r_sq = LinReg.score(X, Y)
print('coefficient of determination:', r_sq)

# Print the intercept of the regression line
# The intercept is the expected mean value of Y when all X=0
print('intercept:', LinReg.intercept_)

# Print the slope (coefficient) of the regression line
# The slope indicates the change in Y for a one-unit change in X
print('slope:', LinReg.coef_)
```
